Este script ha sido creado con el fin de generar una matriz (en este caso para matk) con todas las secuencias que se tiene por genero. Por ello se quizo eliminar todo lo sobrante en el nombre y quedarse solo con el vial_code. Y además comprobar que las muestras existiesen en el excel.

```python
##### eliminar lo que hay despues del "_"

# Definir una lista para almacenar las líneas modificadas
modified_lines = []

# Definir una función para procesar cada línea del archivo
def process_line(line):
    # Verificar si la línea comienza con '>'
    if line.startswith('>'):
        # Extraer la información antes del primer "_"
        seq_id = line[1:].split('_', 1)[0]  # Obtener la parte del identificador antes del primer guion bajo
        # Verificar si el identificador ha sido modificado
        if seq_id != line[1:].strip():  # Si la parte antes del "_" es diferente a la línea original sin espacios
            modified_lines.append(f">{seq_id}")  # Agregar el identificador modificado a la lista
        return f">{seq_id}"  # Crear un nuevo encabezado con la parte del identificador antes del primer guion bajo
    else:
        return line  # Devolver la línea sin cambios si no comienza con '>'

# Rutas de los archivos de entrada y salida
input_file_path = "/content/lobularia_rox_final_modificado.fas"  # Ruta del archivo de entrada FASTA
output_file_path = "campylanthus.fasta"  # Ruta del archivo de salida FASTA

# Abrir el archivo de entrada en modo lectura
with open(input_file_path, 'r') as input_file, open(output_file_path, "w") as output_file:
    # Iterar sobre cada línea del archivo de entrada
    for line in input_file:
        # Procesar cada línea
        processed_line = process_line(line.strip())
        output_file.write(processed_line + '\n')  # Escribir la línea procesada en el archivo de salida

mod_line = [line.strip(">") for line in modified_lines]
print("\nLista modificada", mod_line, "Longitud:", len(mod_line))
```

```python
#@@@@@@@@ comparar excel y mi lista de modificados
import pandas as pd

# Leer el archivo de Excel y obtener la primera columna
excel_file_path = "/content/REGISTRO DNABANK_NGD_2023_metadatos.xlsx"  # Ruta del archivo de Excel

# Leer el archivo Excel y obtener la primera columna
excel_data = pd.read_excel(excel_file_path, header=None)
target_column = excel_data.iloc[:, 0].astype(str)  # Obtener la primera columna y convertirla a tipo str para asegurar la comparación

# Definir una lista para almacenar los resultados de la comparación
results = []

# Organizar las líneas modificadas y los resultados por "encontradas" y "no encontradas"
found_results = []
not_found_results = []

# Iterar sobre las líneas modificadas
for modified_line in mod_line:
    # Verificar si la línea modificada está contenida en la primera columna del archivo Excel
    if modified_line in target_column.values:
        result = "Encontrado"
        found_results.append((modified_line, result))
    else:
        result = "No encontrado"
        not_found_results.append((modified_line, result))

    # Almacenar el resultado
    results.append(result)

# Imprimir los resultados organizados
print("Resultados encontrados:")
for modified_line, result in found_results:
    print(f"Línea modificada: {modified_line} - Resultado: {result}", )
print("\nNumero de encontrados:", len(found_results))

print("\nResultados no encontrados:")
for modified_line, result in not_found_results:
    print(f"Línea modificada: {modified_line} - Resultado: {result}")
print("\nNumero de no encontrados:", len(not_found_results))
```

```python
##### COMPARAR ARCHIVO DIRECTAMENTE, cuando no hemos modificado nada
import pandas as pd

# Ruta del archivo de Excel
excel_file_path = "/content/REGISTRO DNABANK_NGD_2023_metadatos.xlsx"

# Leer el archivo Excel y obtener la primera columna
excel_data = pd.read_excel(excel_file_path, header=None)
target_column = excel_data.iloc[:, 0].astype(str)

# Ruta del archivo con las líneas a comparar
archivo_a_comparar = "/content/MATRIZ_MATK_R.fas"  # Reemplaza "ruta_del_archivo.txt" con la ruta real de tu archivo

# Leer las líneas del archivo a comparar
with open(archivo_a_comparar, 'r') as file:
    lineas_a_comparar = file.readlines()

# Definir listas para almacenar los resultados de la comparación
found_results = []
not_found_results = []

# Iterar sobre las líneas a comparar
for line in lineas_a_comparar:
    # Verificar si la línea comienza con '>'
    if line.startswith('>'):
        # Obtener la parte de la línea que se comparará con el archivo Excel (sin el '> ' al principio y sin el salto de línea al final)
        modified_line = line.strip()[1:]

        # Verificar si la línea modificada está contenida en la primera columna del archivo Excel
        if modified_line in target_column.values:
            result = "Encontrado"
            found_results.append((modified_line, result))
        else:
            result = "No encontrado"
            not_found_results.append((modified_line, result))

# Imprimir los resultados organizados
print("Resultados encontrados:")
for modified_line, result in found_results:
    print(f"Línea encontrada: {modified_line} - Resultado: {result}")
print("\nNumero de encontrados:", len(found_results))

print("\nResultados no encontrados:")
for modified_line, result in not_found_results:
    print(f"Línea no encontrada: {modified_line} - Resultado: {result}")
print("\nNumero de no encontrados:", len(not_found_results))
```

```python
####### generar txt con los identificadores si encontrados y con los que no.
found_output_file = "resultados_encontrados.txt"
not_found_output_file = "resultados_no_encontrados.txt"

# Escribir resultados encontrados en un archivo de texto
with open(found_output_file, 'w') as found_file:
    found_file.write("Resultados encontrados:\n")
    for modified_line, result in found_results:
        found_file.write(f"Línea encontrada: {modified_line} - Resultado: {result}\n")
    found_file.write(f"\nNumero de encontrados: {len(found_results)}\n")

# Escribir resultados no encontrados en un archivo de texto
with open(not_found_output_file, 'w') as not_found_file:
    not_found_file.write("Resultados no encontrados:\n")
    for modified_line, result in not_found_results:
        not_found_file.write(f"Línea no encontrada: {modified_line} - Resultado: {result}\n")
    not_found_file.write(f"\nNumero de no encontrados: {len(not_found_results)}\n")
```
